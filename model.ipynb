```python
# Discussion on which columns to keep and whom to remove and why and how 

# + genres : cuz movies are cateorized based on the genres
# + id : to fetch images while builing the website
# + keywords : to make a matching tags between the script of the movies
# - language : people prefer language but still we have no good distribution of data as 95% is in english in and remaining in other languages
# + title : original and titles are same 
# - original title: as both title and original title is same but since there are some regional movies there title might mein that language
# + overview : since the same and matching content movies may someone like so for that we are going to keep overview
# - popularity : this can be a good measure to recommend as of trending movies but right now we are not making a commercial movie reviewer 
# - production companies : makes no sense
# - production countries : again makes no sense
# - release date : again makes sense but not making for commercial purpose
# - revenue : again as above
# - runtime :
# - spoken language :
# - status :
# - tagline : 
# - vote average : 
# - vote count : 
# - movie_id :
# + cast : since people are fans of some actors
# + crew : as directors and producers are also involved

movies = movies[['movie_id','genres','keywords','title','overview','cast','crew']]
```

```python
# movie_id
# title 
# tags = overview + generes + keywords + cast(main lead) + crew(director)
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def convert(obj):
    obj = ast.literal_eval(obj)
    lst =[]
    for i in obj:
        lst.append(i["name"])
    return lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert1(obj):
    lst =[]
    counter = 0
    for i in ast.literal_eval(obj):
        counter += 1
        if counter <= 3:
            lst.append(i["name"])
        else:
            break
    return lst


movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies.head(1)
```

```python
def convert2(obj):
    lst =[]
    for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                lst.append(i["name"])
    return lst


movies['crew'] = movies['crew'].apply(convert2)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head(1)
```

```python
# 'Sam Worthington' => 'SamWorthington'
```

```python
movies['genres'] = movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies["overview"] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
df = movies[['movie_id','title','tags']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : " ".join(x))
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : x.lower())
```

```python
df.head()
```

```python
# The ultimate goal is the that is tags of 2 rows are same that means there values are and can be recommended so to ultimately we need to find a realtionship between tags of all the rows 

# Now can we do that

# Method 1: we can make or find number of similar words among the tags 
# Method 2: we can vectorize them by giving a value in a 3 dimensional space  known as vectorization 

# The technique we are using is called `bag of words`
```

```python
# from the whole set of those words we form a table like structure

# word0 | word1 | word2 | word3 | word4 | word5 | word6 and so on 
# word1
# word2
# word3 
# word4
# word5
# word6
# and so onn.

```

```python
# great you got till here 

# but now you can see there are few words that are repeating in plural form like activity & activties and action and actions 
# to overcome this we are using stemming 

# ['loved','loving','lover'] => ['love','love','love']

# To overcome this we will use a library called nltk
```

```python
#%pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
ps.stem('actions')
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
# this is not mistakenly repetited it is intentional 
# now observe this time that you are getting term only root ones
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()[688:900]
```

```python
# now we are going to calculate the distance between one vector will all other vectors and vice-versa
# secondly we are not going to calculate the euclidean distance we are going to calculate the cosine distance between them 

# the more the distance will the faulty the euclidiean distance so that is why we will use angle 
# distance inversly proportional to similarity 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x : x[1])[1:6]
```

```python
sorted(similarity[0],reverse=True)
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x : x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]].title)
    
```

```python
recommend("Avatar")
```
